### INITIALIZE THE MODEL WITH GROQ PROVIDER

```python
from langchain.chat_models import init_chat_model
import os

# Set your Groq API key
os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")

# Initialize the model with provider
model = init_chat_model(
    "qwen/qwen3-32b",
    model_provider="groq"  # <-- explicitly set provider
)

# Invoke the model
response = model.invoke("Why do parrots talk?")
print(response.content)
```

```python
from langchain_groq import ChatGroq

llm = ChatGroq(
    model="qwen/qwen3-32b",
    temperature=0.2,
    top_p=0.9,
    frequency_penalty=0.3,
    presence_penalty=0.2,
    reasoning_format="raw",
    streaming=True,
    verbose=True
)
```

### 1ï¸âƒ£ What is Hallucination?

Hallucination = model confidently saying wrong / made-up info

Example:

âŒ â€œPython was created in 1998 by Elon Muskâ€

Why it happens:

Model doesnâ€™t know the answer

No context / weak prompt

Forced to answer (temperature too high)

No grounding (docs, tools, retrieval)

```python
from langchain_groq import ChatGroq

llm = ChatGroq(
    model="qwen/qwen3-32b",
    temperature=2  # âŒ high randomness
)

res = llm.invoke("Who invented Python in 2005?")
print(res.content)
```

```python
from sympy import pretty_print
context = """
Python was created by Guido van Rossum and first released in 1991.
"""

prompt = f"""
Answer ONLY using the context below.
If answer is not in context, say "Not mentioned".

Context:
{context}

Question:
Who invented Python in 2005?
"""

llm = ChatGroq(
    model="qwen/qwen3-32b",
    temperature=0
)

mode=pretty_print(llm.invoke(prompt).content)
mode

```

### 4ï¸âƒ£ What is batch()?

Multiple prompts at once (faster & cheaper)

```python
questions = [
    "What is Python?",
    "Who created Linux?",
    "What is Java?"
]

responses = llm.batch(questions)

for r in responses:
    print(r.content)
```

```python
llm.batch(
    ["Why do parrots have colorful feathers?",
    "How do airplanes fly?",
    "What is quantum computing?"],
    config={
        'max_concurrency': 5,  # Limit to 5 parallel calls
    }
)
```

### 5ï¸âƒ£ What is STREAMING?

Get tokens as they are generated (real-time output)

```python

from langchain_groq import ChatGroq

llm = ChatGroq(
    model="qwen/qwen3-32b",
    temperature=0,
    streaming=True,   # for streaming parameters
)

# stream() returns an iterator
for chunk in llm.stream("Explain hallucination in LLMs in simple terms."):
    # Each chunk is an AIMessageChunk
    if chunk.content:
        print(chunk.content, end="", flush=True)
```
